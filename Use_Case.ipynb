## Use case for the QCDC

Imports

```python
from qiskit import *
from QCDC import print_submatrices
import numpy as np
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
```

The Quantum Circuit for our case

```python
qc = QuantumCircuit(3)

qc.h([0,1,2])
qc.barrier()

qc.cx(2,1)
qc.barrier()

qc.y(1)
qc.x(0)
qc.barrier()

qc.z(2)
qc.x([0,1])
qc.barrier()

qc.cswap(2,1,0)
```

Draw Circuit

```python
qc.draw('mpl',reverse_bits=True)
```

Call the print_submatrices() function to print the operator matrices at each step.

```python
print_submatrices(qc=qc, display_circ=True)
```

Apply measurement

```python
qc.measure_all() #Add measurement
```

Run circuit and plot the corresponding probabilities histogram.

```python
simulator = AerSimulator()
result = simulator.run(qc, shots=1024).result()
counts = result.get_counts()

print(counts)
plot_histogram(counts)
```
